```python
import pandas as pd
import numpy as np
import time
from tqdm import tqdm
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
```

```python
merged_df = pd.read_csv("JOBS_DATA.csv")
```

## Taking necessary details from a job posting

```python
chrome_path = r'chrome_driver_path'
service = Service(chrome_path)
options = Options()
user_agent = "your_user_agent"
options.add_argument(f'user-agent={user_agent}')
browser = webdriver.Chrome(service=service, options=options)

data2 = []

for link in tqdm(merged_df["link"]):
    
    browser.get(link)
    time.sleep(3)
    soup = BeautifulSoup(browser.page_source, "html.parser")
    
    try:
        openings = soup.find("div", class_="styles_jhc__jd-stats__KrId0").find_all("span")[3].text
    except:
        openings = np.nan
    try:
        applicants = soup.find("div", class_="styles_jhc__jd-stats__KrId0").find_all("span")[4].find("span").text
    except:
        applicants = np.nan
    try:
        industry_type = soup.find("div", class_="styles_other-details__oEN4O").find_all("span")[2].text
    except:
        industry_type = np.nan
    try:
        employment_type = ", ".join(emp_type.text.strip() for emp_type in soup.find("div", class_="styles_other-details__oEN4O").find_all("span")[6])
    except:
        employment_type = np.nan
    try:
        degree = soup.find("div", class_="styles_education__KXFkO").find_all("div")[1].text[4:]
    except:
        degree = np.nan
    try:
        specialization = soup.find("div", class_="styles_education__KXFkO").find_all("div")[2].text[4:]
    except:
        specialization = np.nan

    data2.append([openings, applicants, industry_type, employment_type, degree, specialization])

print(len(data2))
```

```python
df_2 = pd.DataFrame(data=data2, columns=["job_openings","applicants","industry_type","employment_type","degree_required","specialization_required"])
```

```python
df_2.to_csv("Details_Job.csv",index=False)
```

## Handling null values

```python
df.dropna(subset=["location"],inplace=True)
df["required_experience"].fillna("0-1", inplace=True)
df["skills"].fillna("Not Specified", inplace=True)
df.dropna(subset=["job_openings"], inplace=True)
df["applicants"].fillna("0", inplace=True)
df["specialization_required"].fillna("Not Specified", inplace=True)
```

```python
df.isnull().sum()
```

```python
for i in range(876, 878):
    val = str(df.loc[i, "job_openings"]).strip()
    parts = val.split()
    if len(parts) >= 3:
        df.loc[i, "job_openings"] = parts[2]  # '10' from 'Less than 10'
    else:
        df.loc[i, "job_openings"] = val  # keep as is
```

```python
df["job_openings"] = df["job_openings"].astype("int")
df["applicants"] = df["job_openings"].astype("int")
df.info()
```
